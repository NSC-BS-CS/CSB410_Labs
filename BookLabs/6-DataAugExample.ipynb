```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

# Load sample image from MNIST
(x_train, y_train), (_, _) = mnist.load_data()
x_sample = x_train[0]  # choose the first image
x_sample = x_sample.reshape((28, 28, 1))  # reshape for ImageDataGenerator
x_sample = np.expand_dims(x_sample, axis=0)  # add batch dimension

# Create ImageDataGenerator with augmentation
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True  # more useful in other domains
)

# Generate augmented images
augmented_images = [next(datagen.flow(x_sample, batch_size=1))[0].reshape(28, 28) for _ in range(9)]

# Plot original + augmented images
fig, axes = plt.subplots(1, 10, figsize=(15, 3))
axes[0].imshow(x_sample[0].reshape(28, 28), cmap='gray')
axes[0].set_title("Original")
axes[0].axis('off')
for i in range(9):
    axes[i+1].imshow(augmented_images[i], cmap='gray')
    axes[i+1].axis('off')
plt.tight_layout()
plt.show()
```
